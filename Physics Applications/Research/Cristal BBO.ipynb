# Índices Ordinario y Extraordinario para cristal BBO

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
nechen_pd = pd.read_csv("BBOChen-e.csv")
nochen_pd = pd.read_csv("BBOChen-o.csv")
neeimerl_pd = pd.read_csv("BBOEimerl-e.csv")
noeimerl_pd = pd.read_csv("BBOEimerl-o.csv")
```

```python
wl1 = np.linspace(0.188,5.2,101)
```

## Tamošauskas (2018)

$$n_e^2-1=\frac{1.151075λ^2}{λ^2-0.007142}+\frac{0.21803λ^2}{λ^2-0.02259}+\frac{0.656λ^2}{λ^2-263}$$


$$n_o^2-1=\frac{0.90291λ^2}{λ^2-0.003926}+\frac{0.83155λ^2}{λ^2-0.018786}+\frac{0.76536λ^2}{λ^2-60.01}$$

```python
ne2018_eq = (1+1.151075/(1-0.007142/wl1**2)+0.21803/(1-0.02259/wl1**2)+0.656/(1-263/wl1**2))**.5
no2018_eq = (1+0.90291/(1-0.003926/wl1**2)+0.83155/(1-0.018786/wl1**2)+0.76536/(1-60.01/wl1**2))**.5
```

## Chen (2012)

$$n_e^2=2.3698703+\frac{0.0128445}{\lambda^2-0.0153064}-0.0029129\lambda^2$$


$$n_o^2=2.7366122+\frac{0.0185720}{\lambda^2-0.0178746}-0.0143756\lambda^2$$

```python
ne2012_eq = (2.3698703+0.0128445/(wl1**2-0.0153064)-0.0029129*wl1**2)**.5
no2012_eq = (2.7366122+0.0185720/(wl1**2-0.0178746)-0.0143756*wl1**2)**.5
```

### Extraordinary Index

```python
nefig2018 = plt.figure(figsize=(15,7.5))
ax2018e = nefig2018.add_subplot(111)
ax2018e.plot(wl1,ne2018_eq,'g',label='Dispersion Formula 2018')
ax2018e.plot(wl1,ne2012_eq,'r',label='Dispersion Formula 2012')
ax2018e.scatter(nechen_pd.wvl,nechen_pd.nex,c='b',label="Chen Data")
ax2018e.scatter(neeimerl_pd.wvl,neeimerl_pd.nex,c='k',marker='+',label="Eimerl Data")
ax2018e.set_title('Indice extraordinario')
ax2018e.set_xlabel(r'Longitud de onda ($\mu$m)')
ax2018e.set_ylabel(r'$n_e$: índice de refracción extraordinario')
ax2018e.legend()
plt.show()
```

### Ordinary Index

```python
nofig2018 = plt.figure(figsize=(15,7.5))
ax2018o = nofig2018.add_subplot(111)
ax2018o.plot(wl1,no2018_eq,'g',label='Dispersion Formula 2018')
ax2018o.plot(wl1,no2012_eq,'r',label='Dispersion Formula 2012')
ax2018o.scatter(nochen_pd.wvl,nochen_pd.no,c='b',label="Chen Data")
ax2018o.scatter(noeimerl_pd.wvl,noeimerl_pd.no,c='k',marker='+',label="Eimerl Data")
ax2018o.set_title('Indice ordinario')
ax2018o.set_xlabel(r'Longitud de onda ($\mu$m)')
ax2018o.set_ylabel(r'$n_o$: índice de refracción ordinario')
ax2018o.legend()
plt.show()
```

# Otras fórmulas de Dispersión

## Zhang (2000): Dispersion Formula

```python
wl2 = np.linspace(0.64,3.18,101)
```

$$n_e^2=2.3753+\frac{0.01224}{λ^2-0.01667}-0.01627λ^{2}+0.0005716λ^{4}-0.00006305λ^{6}$$


$$n_o^2=2.7359+\frac{0.01878}{λ^2-0.01822}-0.01471λ^{2}+0.0006081λ^{4}-0.00006740λ^{6}$$

```python
ne2000_eq = (2.3753+0.01224/(wl2**2-0.01667)-0.01627*wl2**2+0.0005716*wl2**4-0.00006305*wl2**6)**.5
no2000_eq = (2.7359+0.01878/(wl2**2-0.01822)-0.01471*wl2**2+0.0006081*wl2**4-0.00006740*wl2**6)**.5
```

```python
nefig2000 = plt.figure(figsize=(15,7.5))
ax2000e = nefig2000.add_subplot(111)
ax2000e.plot(wl2,ne2000_eq,'c',label='Dispersion Formula 2000')
ax2000e.scatter(nechen_pd.wvl,nechen_pd.nex,c='b',label="Chen Data")
ax2000e.scatter(neeimerl_pd.wvl,neeimerl_pd.nex,c='k',marker='+',label="Eimerl Data")
ax2000e.set_title('Indice extraordinario (2000)')
ax2000e.set_xlabel(r'Longitud de onda ($\mu$m)')
ax2000e.set_ylabel('n: índice de refracción')
ax2000e.legend()
plt.show()
```

```python
nofig2000 = plt.figure(figsize=(15,7.5))
ax2000o = nofig2000.add_subplot(111)
ax2000o.plot(wl2,no2000_eq,'c',label='Dispersion Formula')
ax2000o.scatter(nochen_pd.wvl,nochen_pd.no,c='b',label="Chen Data")
ax2000o.scatter(noeimerl_pd.wvl,noeimerl_pd.no,c='k',marker='+',label="Eimerl Data")
ax2000o.set_title('Indice ordinario (2000)')
ax2000o.set_xlabel(r'Longitud de onda ($\mu$m)')
ax2000o.set_ylabel('n: índice de refracción')
ax2000o.legend()
plt.show()
```
