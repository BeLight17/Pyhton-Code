# $LiIO_3$

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Herbst (1976)

```python
nelio_pd = pd.read_csv("LiIO3Herbst-e.csv")
nolio_pd = pd.read_csv("LiIO3Herbst-o.csv")
```

```python
nelio_pd
```

```python
wl1 = np.linspace(0.546,5.0,101)
```

$$n_e^2-1=0.83086+\frac{1.08807λ^2}{λ^2-0.0313810}+\frac{0.554582λ^2}{λ^2-158.76}$$


$$n_o^2-1=1.03132+\frac{1.37623λ^2}{λ^2-0.0350832}+\frac{1.06745λ^2}{λ^2-169.00}$$

```python
nelio_eq = (1+0.83086+1.08807/(1-0.0313810/wl1**2)+0.554582/(1-158.76/wl1**2))**.5
nolio_eq = (1+1.03132+1.37623/(1-0.0350832/wl1**2)+1.06745/(1-169.00/wl1**2))**.5
```

```python
nefiglio = plt.figure(figsize=(15,7.5))
axlioe = nefiglio.add_subplot(111)
axlioe.plot(wl1,nelio_eq,'c',label='Dispersion Formula 1976')
axlioe.scatter(nelio_pd.wvl,nelio_pd.nex)
axlioe.set_title(r'Indice extraordinario $LiIO_3$')
axlioe.set_xlabel(r'Longitud de onda ($\mu$m)')
axlioe.set_ylabel(r'$n_e$: índice de refracción extraordinario')
axlioe.legend()
plt.show()
```

```python
nofiglio = plt.figure(figsize=(15,7.5))
axlioo = nofiglio.add_subplot(111)
axlioo.plot(wl1,nolio_eq,'c',label='Dispersion Formula 1976')
axlioo.scatter(nolio_pd.wvl,nolio_pd.no)
axlioo.set_title(r'Indice ordinario $LiIO_3$')
axlioo.set_xlabel(r'Longitud de onda ($\mu$m)')
axlioo.set_ylabel(r'$n_o$: índice de refracción ordinario')
axlioo.legend()
plt.show()
```

# KBBF

## Chen (2012)

```python
nekbbf_pd = pd.read_csv("KBBFChen-e.csv")
nokbbf_pd = pd.read_csv("KBBFChen-o.csv")
```

```python
nekbbf_pd
```

```python
wl2 = np.linspace(0.4047,0.6562,101)
```

$$n_e^2=1+\frac{0.9316λ^2}{λ^2-0.00675}-0.00169λ^2$$


$$n_o^2=1+\frac{1.1713λ^2}{λ^2-0.00733}-0.01022λ^2$$

```python
nekbbf_eq = (1+0.9316*wl2**2/(wl2**2-0.00675)-0.00169*wl2**2)**.5
nokbbf_eq = (1+1.1713*wl2**2/(wl2**2-0.00733)-0.01022*wl2**2)**.5
```

```python
nefigkbbf = plt.figure(figsize=(15,7.5))
axkbbfe = nefigkbbf.add_subplot(111)
axkbbfe.plot(wl2,nekbbf_eq,'y',label='Dispersion Formula 2012')
axkbbfe.scatter(nekbbf_pd.wvl,nekbbf_pd.nex,c='g')
axkbbfe.set_title(r'Indice extraordinario KBBF')
axkbbfe.set_xlabel(r'Longitud de onda ($\mu$m)')
axkbbfe.set_ylabel(r'$n_e$: índice de refracción extraordinario')
axkbbfe.legend()
plt.show()
```

# CBBF

## Chen (2012)

```python
necbbf_pd = pd.read_csv("CBBFChen-e.csv")
nocbbf_pd = pd.read_csv("CBBFChen-o.csv")
```

```python
necbbf_pd
```

```python
wl3 = np.linspace(0.2537,1.014,101)
```

$$n_e^2=2.0802682+\frac{0.0070027}{λ^2-0.0109331}-0.0047423λ^2$$


$$n_o^2=2.2562126+\frac{0.0091453}{λ^2-0.0126509}-0.0101828λ^2$$

```python
necbbf_eq = (2.0802682 + 0.0070027/(wl3**2 - 0.0109331) - 0.0047423*wl3**2)**0.5
nocbbf_eq = (2.2562126 + 0.0091453/(wl3**2 - 0.0126509) - 0.0101828*wl3**2)**0.5
```

```python
nefigcbbf = plt.figure(figsize=(15,7.5))
axcbbfe = nefigcbbf.add_subplot(111)
axcbbfe.plot(wl3,necbbf_eq,'m',label='Dispersion Formula 2012')
axcbbfe.scatter(necbbf_pd.wvl,necbbf_pd.nex,c='r')
axcbbfe.set_title(r'Indice extraordinario CBBF')
axcbbfe.set_xlabel(r'Longitud de onda ($\mu$m)')
axcbbfe.set_ylabel(r'$n_e$: índice de refracción extraordinario')
axcbbfe.legend()
plt.show()
```

```python
nofigcbbf = plt.figure(figsize=(15,7.5))
axcbbfo = nofigcbbf.add_subplot(111)
axcbbfo.plot(wl3,nocbbf_eq,'m',label='Dispersion Formula 2012')
axcbbfo.scatter(nocbbf_pd.wvl,nocbbf_pd.no,c='r')
axcbbfo.set_title(r'Indice ordinario CBBF')
axcbbfo.set_xlabel(r'Longitud de onda ($\mu$m)')
axcbbfo.set_ylabel(r'$n_e$: índice de refracción ordinario')
axcbbfo.legend()
plt.show()
```
