```python
import matplotlib.pyplot as plt
import numpy as np
from math import *
```

```python
def plot(fig,ax,x,y,pars,title):
    ax[0].plot(x,y[0][0],'g-.',label=r"$\gamma(x)$")
    ax[0].plot(x,y[0][1],'b',label=r"$\delta(x)$")
    ax[0].plot(x,y[0][2],'r--',label=r"$\delta'(x)$")
    ax[0].legend()
    ax[0].set_title('Parámetro = ' + pars[0])
    ax[1].plot(x,y[1][0],'g-.',label=r"$\gamma(x)$")
    ax[1].plot(x,y[1][1],'b',label=r"$\delta(x)$")
    ax[1].plot(x,y[1][2],'r--',label=r"$\delta'(x)$")
    ax[1].legend()
    ax[1].set_title('Parámetro = ' + pars[1])
    plt.suptitle(title,fontsize=15,position = (0.5,0.97))
    fig.savefig(title+'.pdf')
    plt.show()
```

```python
def agamma(x):
    if x < 0:
        return -1/2
    else:
        return 1/2
    
def adelta(x,a):
    return a/(np.pi*(a**2+x**2))

def adelta_p(x,a):
    return -2*a*x/(np.pi*(x**2+a**2)**2)
```

```python
def bgamma(phi,r):
    return (np.arctan((r+1)*np.tan(phi/2)/(1-r)))/np.pi
    
def bdelta(phi,r):
    return (1-r**2)/(2*pi*(1-2*r*np.cos(phi)+r**2))

def bdelta_p(phi,r):
    return -r*(1-r**2)*np.sin(phi)/(pi*(1-2*r*np.cos(phi)+r**2)**2)
```

```python
def cgamma(x,t):
    return 1/2 + np.arctan(x/t)/np.pi
    
def cdelta(x,t):
    return t/(np.pi*(x**2+t**2))

def cdelta_p(x,t):
    return -2*t*x/(np.pi*(x**2+t**2)**2)
```

```python
x = np.linspace(-5,5,500)
y11 = np.array([agamma(x) for x in x])
y12 = np.array([adelta(x,0.5) for x in x])
y13 = np.array([adelta_p(x,0.5) for x in x])
y21 = y11
y22 = np.array([adelta(x,0.2) for x in x])
y23 = np.array([adelta_p(x,0.2) for x in x])
```

```python
fig, ax = plt.subplots(1,2,figsize=(8.3,3.9))
plot(fig,ax,x,[[y11,y12,y13],[y21,y22,y23]],['0.5','0.2'],'Pregunta 2a')
```
