# Secant Method

A partir de dos puntos $a, b$ se determina una recta que interseca al eje X en el punto $x$, donde $$x = \frac{af(b)-bf(a)}{f(b)-f(a)}$$

Luego se toman como nuevos extremos el punto $x$ y el más cercano de los puntos iniciales, y el proceso se repite hasta que la distancia entre dos puntos sea menor a un valor denominado *tolerance*.

```python
from math import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definimos la función f(x)
f = lambda x: cos(x**2)*sin(x)
```

```python
# Graficamos la función en un intervalo
li = -pi
ls = pi
x_arr = np. linspace(li,ls,150)
y_arr = np.array([f(x) for x in x_arr])
```

```python
fig, ax = plt.subplots(1,1,figsize=(15,10))
ax.plot(x_arr,y_arr,'g')
ax.axhline(color='k')
plt.show()
```

```python
def secmet(f,a,b,tolerance=1.0e-6):
    dx = abs(a-b)
    
    while dx > tolerance:
        x = (a*f(b) - b*f(a))/(f(b) - f(a))
        
        if abs(x-a) < abs(x-b):
            b = x
        else:
            a = x
        
        dx = abs(a-b)
        
    print("Found f(x) = 0 at %.8f +/- %.8f"%(a,tolerance))
```

```python
secmet(f,-1.5,-1)
```
