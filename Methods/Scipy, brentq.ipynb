```python
from math import *
from scipy import optimize
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definimos la función como una función de Python
def f(x):
    return cos(x**2)*sin(x)
```

```python
# Graficamos la función

li = -pi
ls = pi
x_arr = np. linspace(li,ls,150)
y_arr = np.array([f(x) for x in x_arr])

fig, ax = plt.subplots(1,1,figsize=(15,10))
ax.plot(x_arr,y_arr,'purple')
ax.axhline(color='k')
plt.show()
```

```python
"""En la función brentq se ingresan los parámetros 
    brentq(f: función
           a, b: intervalo en el que se encuentra la raíz
           xtol = 2e-12: tolerance)""" 

sol = optimize.brentq(f,-1.5,-1)

print("Found f(x) = 0 at %.8f"%(sol))
```
